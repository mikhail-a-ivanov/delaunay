```python
from delaunay import PointCloud, Triangulation
```

```python
P = PointCloud(20, min_distance=0.1)
```

```python
P.drawConvexHull(pngname="convex.png")
```

```python
T = Triangulation(P)
```

```python
T.drawTriangulation(pngname="earclipping.png")
```

```python
T.makeDelaunay()
```

```python
T.drawTriangulation(pngname="delaunay.png")
```

```python
T.drawTriangulation(pngname="delaunay-circles.png", showCircumcircles=True)
```
